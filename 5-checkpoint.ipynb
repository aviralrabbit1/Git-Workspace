```python
import numpy as np
import matplotlib.pyplot as plt

#matrix A
m = [[12, 7, 3],
    [4, 5, 6],
    [7, 8, 9]]
  
#matrix B
n = [[5, 8, 1, 2],
    [6, 7, 3, 0],
    [4, 5, 9, 1]]
  
def multiply_without(A,B):
    result1 = [[sum(a * b for a, b in zip(A_row, B_col)) for B_col in zip(*B)]for A_row in A]
    return result1
print(multiply_without(m,n))

#matrix A
o = np.array([[12, 7, 3],
    [4, 5, 6],
    [7, 8, 9]])
  
#matrix B
p = np.array([[5, 8, 1, 2],
    [6, 7, 3, 0],
    [4, 5, 9, 1]])

def multiply_with(A,B):
    result = np.dot(A,B)
    return result
print(multiply_with(o,p))

%timeit multiply_with(o,p)
%timeit multiply_without(m,n)
```
